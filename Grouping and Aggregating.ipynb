```python
country_respondents = df_public['Country'].value_counts()
country_respondents
```

```python
country_group5 = df_public.groupby(['Country'])
country_uses_python = country_group5['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort=False)
python_df
```

```python
python_df.rename(columns = {'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumknowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumknowsPython']/python_df['NumRespondents']) * 100
python_df
```
